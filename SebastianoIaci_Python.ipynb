```python
import re

class Ricetta:
    def __init__(self, nome_ricetta, ingredienti, tempo_preparazione):
        self.nome_ricetta = nome_ricetta
        self.ingredienti = ingredienti
        self.tempo_preparazione = tempo_preparazione

    def stampa_ricetta(self):
        print(f"{self.nome_ricetta}: {', '.join(self.ingredienti)} , {self.tempo_preparazione} minuti")

    def modifica_ricetta(self):
        # Modifica il nome
        nuovo_nome = input(f"Modifica il nome della ricetta (lascia vuoto per mantenere '{self.nome_ricetta}'): ").strip()
        if nuovo_nome:
            if not is_with_number(nuovo_nome):
                self.nome_ricetta = nuovo_nome
            else:
                print("Nome non valido (non pu√≤ contenere numeri), si mantiene il nome precedente.")

        # Modifica gli ingredienti
        modifica_ingredienti = input("Vuoi modificare gli ingredienti? (s/n): ").strip().lower()
        if modifica_ingredienti == 's':
            self.ingredienti = aggiungi_ingredienti()  # Usa la funzione esistente per aggiungere nuovi ingredienti

        # Modifica il tempo di preparazione
        nuovo_tempo = input(f"Modifica il tempo di preparazione (lascia vuoto per mantenere '{self.tempo_preparazione}' minuti): ").strip()
        if nuovo_tempo:
            try:
                self.tempo_preparazione = int(nuovo_tempo)
            except ValueError:
                print("Tempo non valido, mantenendo il valore precedente.")

def is_with_number(stringa):
    return bool(re.search(r'\d', stringa))

def aggiungi_ingredienti():
    lista = []
    while True:
        aggiungi = input("Inserisci un ingrediente, premi 'quit' per chiudere: ")
        if aggiungi.lower() != "quit":
            lista.append(aggiungi)
            print(lista)
        else:
            return lista

def crea_nuova_ricetta():
    nome_ricetta = input("Scrivi il nome di una ricetta: ")
    if is_with_number(nome_ricetta):
        print("Caratteri numerici non ammessi")
        return False

    ingredienti = aggiungi_ingredienti()
    tempo_preparazione = input("Scrivi il tempo di preparazione in minuti: ")

    nuova_ricetta = Ricetta(nome_ricetta, ingredienti, tempo_preparazione)
    return nuova_ricetta

# Funzione per cercare una ricetta in base a un ingrediente
def cerca_ingrediente(ingrediente):
    trovate = False
    for ricetta in lista_ricette:
        if ingrediente in ricetta.ingredienti:
            ricetta.stampa_ricetta()
            trovate = True
    if not trovate:
        print(f"Nessuna ricetta trovata con l'ingrediente '{ingrediente}'.")

# Funzione per cercare ricette in base al tempo di preparazione con operatori
def cerca_tempo_preparazione(condizione):
    trovate = False
    operatore = condizione[0]
    try:
        tempo = int(condizione[1:].strip())  # Rimuove l'operatore e converte il resto in numero
    except ValueError:
        print("Formato non valido. Usa <, > o = seguito da un numero (es: >30).")
        return

    for ricetta in lista_ricette:
        if operatore == '>' and int(ricetta.tempo_preparazione) > tempo:
            ricetta.stampa_ricetta()
            trovate = True
        elif operatore == '<' and int(ricetta.tempo_preparazione) < tempo:
            ricetta.stampa_ricetta()
            trovate = True
        elif operatore == '=' and int(ricetta.tempo_preparazione) == tempo:
            ricetta.stampa_ricetta()
            trovate = True

    if not trovate:
        print(f"Nessuna ricetta trovata con tempo di preparazione {operatore} {tempo} minuti.")

# Lista iniziale di ricette
lista_ricette = [
    Ricetta("Spaghetti all'amatriciana",
            ["guanciale", "olio d'oliva", "vino bianco secco", "pomodori San Marzano", "peperoncino", "pecorino"], 30),
    Ricetta("Caponata siciliana",
            ["salsa di pomodoro", "olive denocciolate", "capperi dissalati", "aceto di vino bianco", "zucchero di canna", "cipolla dorata", "sedano", "olio d'oliva", "basilico", "sale"], 40),
    Ricetta("Fegato alla veneziana",
            ["fegato di vitello", "cipolle bianche", "burro", "vino bianco", "foglia di alloro", "olio extravergine di oliva", "sale", "pepe"], 25),
]

# Funzione per modificare una ricetta esistente
def modifica_ricetta():
    try:
        indice = int(input("Inserisci l'indice della ricetta da modificare: "))
        if indice < len(lista_ricette):
            lista_ricette[indice].modifica_ricetta()
        else:
            print("Indice non valido.")
    except ValueError:
        print("Indice non valido.")

# Loop principale
while True:
    crea_cerca_ricetta = input("Digita 'nuovo' per creare una ricetta, un numero da 0 a 2 per visualizzare una ricetta, un ingrediente per cercarlo nelle ricette, 'tempo' per cercare ricette in base al tempo di preparazione o 'modifica' per modificare una ricetta: ")

    if crea_cerca_ricetta == "nuovo":
        nuova_ricetta = crea_nuova_ricetta()
        if nuova_ricetta:
            lista_ricette.append(nuova_ricetta)
            nuova_ricetta.stampa_ricetta()

    elif crea_cerca_ricetta.isdigit():
        indice = int(crea_cerca_ricetta)
        if indice < len(lista_ricette):
            lista_ricette[indice].stampa_ricetta()
        else:
            print("Indice non valido, riprovare.")

    elif crea_cerca_ricetta.lower() == "tempo":
        condizione = input("Inserisci una condizione di tempo (es: >30, <40, =25): ").strip()
        if condizione[0] in ['>', '<', '='] and len(condizione) > 1:
            cerca_tempo_preparazione(condizione)
        else:
            print("Formato non valido. Usa <, > o = seguito da un numero (es: >30).")

    elif crea_cerca_ricetta.lower() == "modifica":
        modifica_ricetta()

    elif len(crea_cerca_ricetta) > 0:
        cerca_ingrediente(crea_cerca_ricetta)

    else:
        print("Istruzione non valida, riprovare.")
```
